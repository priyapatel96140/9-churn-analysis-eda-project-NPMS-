## Bivariate Analysis

```python
# Does gender affect customer churn?
pd.crosstab(df['gender'], df['Churn'])
round(pd.crosstab(df['gender'], df['Churn'], normalize='index') * 100, 2)
```

```python

plt.figure(figsize=(6,4))

sns.countplot(data=df, x='gender', hue='Churn')

plt.title('Gender vs Customer Churn')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#Do senior citizens, partners, and dependents influence customer churn?
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1,3, figsize=(15,4))

sns.heatmap(pd.crosstab(df['SeniorCitizen'], df['Churn']),
            annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Senior Citizen vs Churn')

sns.heatmap(pd.crosstab(df['Partner'], df['Churn']),
            annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title('Partner vs Churn')

sns.heatmap(pd.crosstab(df['Dependents'], df['Churn']),
            annot=True, fmt='d', cmap='Oranges', ax=axes[2])
axes[2].set_title('Dependents vs Churn')

plt.tight_layout()
plt.show()
```

### output:
Senior Citizens tend to have a higher churn rate than non-senior citizens.
Customers without a partner are more likely to churn than customers with a partner.
Customers without dependents have a higher churn rate than customers with dependents.

### Insight
Senior citizens may need more personalized customer support and retention programs.
Customers without partners or dependents appear to be at a higher risk of churn, so they can be targeted with loyalty offers or special discounts.
Family-oriented customers (those with partners or dependents) generally show better retention, suggesting they may value long-term service stability.

```python
# Does Internet Service affect customer churn?
# Create percentage table
internet_churn = pd.crosstab(df['InternetService'],
                             df['Churn'],
                             normalize='index') * 100

# Plot
internet_churn.plot(kind='bar',
                    stacked=True,
                    figsize=(8,5))

plt.title('Internet Service vs Customer Churn')
plt.xlabel('Internet Service')
plt.ylabel('Percentage of Customers')
plt.legend(title='Churn')
plt.xticks(rotation=0)
plt.show()
```

### output:
Yes, Internet Service affects customer churn. Customers using Fiber Optic have the highest churn rate, while customers with No Internet Service have the lowest churn rate. DSL customers have a moderate churn rate.

### Insight
Fiber Optic customers should be the company's highest priority for retention efforts.
The company should investigate whether higher pricing, service quality, or customer expectations are causing Fiber Optic customers to leave.
Since DSL and No Internet Service customers have lower churn rates, the company can study what contributes to their higher retention and apply similar strategies to Fiber Optic customers.

```python
# Does Contract Type affect customer churn?
# Create percentage table
contract_churn = pd.crosstab(df['Contract'],
                             df['Churn'],
                             normalize='index') * 100

# Plot stacked bar chart
contract_churn.plot(kind='bar',
                    stacked=True,
                    figsize=(8,5))

plt.title('Contract Type vs Customer Churn')
plt.xlabel('Contract Type')
plt.ylabel('Percentage of Customers')
plt.legend(title='Churn')
plt.xticks(rotation=0)
plt.show()
```

### output:
Yes, contract type has a strong effect on customer churn. Customers with month-to-month contracts have the highest churn rate, while customers with two-year contracts have the lowest churn rate. Customers with one-year contracts also churn much less than month-to-month customers.

### Insight
Month-to-month customers are the highest-risk group and should be the primary focus of retention campaigns.
The company should encourage customers to switch to one-year or two-year contracts by offering discounts, loyalty rewards, or additional benefits.
Long-term contracts improve customer retention and provide more stable revenue for the company.

```python
# Does Payment Method affect customer churn?
payment_percent = pd.crosstab(df['PaymentMethod'],
                              df['Churn'],
                              normalize='index') * 100

plt.figure(figsize=(8,5))

sns.heatmap(payment_percent,
            annot=True,
            fmt='.1f',
            cmap='Blues')

plt.title('Payment Method vs Customer Churn (%)')
plt.xlabel('Churn')
plt.ylabel('Payment Method')

plt.show()
```

### output:
Yes, payment method affects customer churn. Customers who use Electronic Check have the highest churn rate, while customers using Automatic Bank Transfer and Automatic Credit Card payments have much lower churn rates.

### Insight
Customers who pay using Electronic Check are at a higher risk of churning.
The company should encourage customers to switch to automatic payment methods by offering incentives such as discounts, cashback, or reward points.
Promoting automatic payments can improve customer convenience, strengthen customer loyalty, and help reduce churn.

```python
# Do Online Security, Online Backup, Device Protection, and Tech Support reduce churn?
services = ['OnlineSecurity', 'OnlineBackup',
            'DeviceProtection', 'TechSupport']

churn_rate = []

for service in services:
    rate = pd.crosstab(df[service], df['Churn'],
                       normalize='index') * 100
    churn_rate.append(rate.loc['Yes', 'Yes'])

plt.figure(figsize=(7,4))

plt.plot(services, churn_rate, marker='o')

plt.title('Churn Rate for Customers Using Different Services')
plt.xlabel('Services')
plt.ylabel('Churn Rate (%)')

plt.xticks(rotation=20)
plt.show()
```

### output:
Yes, value-added services are associated with lower customer churn. Customers who subscribe to Online Security, Online Backup, Device Protection, and Tech Support generally have a lower churn rate than customers who do not subscribe to these services.

### Insight
Customers who subscribe to additional services are generally more loyal.
Customers without these services are at a higher risk of churn.
The company should promote Online Security, Online Backup, Device Protection, and Tech Support through bundled offers or discounts to improve customer retention.
Increasing adoption of these services can help reduce churn and improve long-term customer satisfaction.

```python
# Do Monthly Charges and Tenure influence customer churn?
#Monthly Charges vs Churn check
plt.figure(figsize=(6,4))

sns.violinplot(data=df, x='Churn', y='MonthlyCharges')

plt.title('Monthly Charges vs Customer Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')

plt.show()

#Tenure vs Churn
plt.figure(figsize=(6,4))

sns.violinplot(data=df, x='Churn', y='tenure')

plt.title('Tenure vs Customer Churn')
plt.xlabel('Churn')
plt.ylabel('Tenure (Months)')

plt.show()
```

### output:
Yes, both Monthly Charges and Tenure influence customer churn.

Customers with higher monthly charges are generally more likely to churn.
Customers with shorter tenure are more likely to churn, while customers who have stayed longer with the company are less likely to leave.

### Insight
Customers paying higher monthly charges may feel they are not receiving enough value, increasing the likelihood of churn.
New customers are at greater risk of leaving, so the company should focus on improving their experience during the first few months.
Offering welcome discounts, loyalty rewards, or personalized support to new and high-paying customers can help reduce churn.

```python
# What are the key factors associated with customer churn?

#To summarize the variables that have the strongest relationship with customer churn based on the previous analyses.
# Selecting numerical columns
num_df = df[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']]

# Correlation matrix
corr = num_df.corr()

# Heatmap
plt.figure(figsize=(6,4))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Correlation Between Numerical Variables')

plt.show()
```

Value-added Services – Customers without Online Security, Online Backup, Device Protection, and Tech Support are more likely to churn.
Monthly Charges – Customers with higher monthly charges tend to churn more.
Tenure – Customers with shorter tenure are more likely to churn than long-term customers.

### Insight
